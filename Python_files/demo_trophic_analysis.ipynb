```python
# Get what we need
import trophic_tools as ta # will need to ensure on path
import networkx as nx 
```

Generate, anayse and visualise some example networks:

```python
# Example 1:

G=nx.DiGraph()
G.add_weighted_edges_from([(1,2,1),(2,3,1),(1,3,1)])
ta.trophic_levels(G)
ta.trophic_plot(G,k=1)
F_0, h = ta.trophic_incoherence(G)
print('Trophic incoherence =',round(F_0,3))
```

```python
# Example 2:

G=nx.DiGraph()
G.add_weighted_edges_from([(1,2,1),(3,2,1),(4,2,1),(2,5,1),(2,6,1),(2,7,1)])
ta.trophic_plot(G)
F_0, h = ta.trophic_incoherence(G)
print('Trophic incoherence =',round(F_0,3))
```

```python
# Example 3:

# Read data file and define network
import pandas as pd 
#file_name='/Users/path.../ythan_E.csv' # update file path. ythan_E.csv available for download with Python code
#df = pd.read_csv(file_name)
url='https://github.com/BazilSansom/Trophic-Analysis-Toolbox/blob/master/Python_files/ythan_E.csv?raw=true'
df = pd.read_csv(url)
#df.head(6)
Graphtype = nx.DiGraph()
G_ythan = nx.from_pandas_edgelist(df, source='source', target='target',edge_attr='weight', create_using=Graphtype)
ta.trophic_plot(G_ythan,k=1)
F_0, h = ta.trophic_incoherence(G_ythan)
print('Trophic incoherence =',round(F_0,3))
```
